# Parte 1: Preprocesamiento y Análisis Exploratorio en Python 

### 1. Carga y limpieza de datos 

#### Cargar el dataset en un DataFrame de Pandas

```python
import pandas as pd 

file_path = r"C:\Users\Roger Saavedra\Desktop\Maestria Cienca de Datos\Prueba Trabajo\Data_pueba_candidatos.xlsx"

df = pd.read_excel(file_path)

df.head()
```

```python
# Realizo una copia del df original

df_cleaned = df.copy()
```

#### Eliminar respuestas parciales (status = "Parcial")

```python
# Veo que solo hay 2 valores unicos Completa y Parcial

df_cleaned = df_cleaned[df_cleaned['status'] != "Parcial"]

df_cleaned.nunique()
```

```python
df_cleaned.shape
```

### Primero identifico variables clave antes de manejar valores nulos

#### Normalizar nombres de columnas (sin espacios ni caracteres especiales)

```python
df_cleaned.columns = (
    df_cleaned.columns
    .str.lower()
    .str.replace(' ', '_')
    .str.replace('[^a-z0-9_]', '', regex=True)
)

print(df_cleaned.columns)
```

```python
# Procedo a cambiar los nombres de cada columna de forma manual utilizando el enunciado deL problema
# Que variables puedo identificar como claves al leer el problema?:
# satisfacción general, tipo de visitante, facilidad de acceso, Promedio del tiempo de visita, NPS.


df_cleaned.rename(columns={
    "1_en_una_escala_de_0_a_10_donde_0_es_no_lo_recomendara_y_10_es_seguro_que_lo_recomendara": "nps",
    "5_con_quin_ha_realizado_la_visita": "tipo_visitante",
    "6_en_general_dira_que_cumpli_con_sus_expectativas": "cumplio_expectativas",
    "9_valore_de_0_a_10_los_siguientes_elementos_experiencia_general": "satisfaccion_general",
    "10_facilidad_en_el_proceso_de_compra": "facilidad_compra",
    "11_facilidad_de_acceso_": "facilidad_acceso",
    "12_claridad_del_recorrido": "claridad_recorrido",
    "13_fluidez_y_movilidad_durante_la_visita": "fluidez_visita",
    "14_calidad_de_los_elementos_expuestos": "calidad_elementos",
    "21_variedad_de_los_elementos_expuestos": "variedad_elementos",
    "22_interactividad__tecnologa_de_las_exhibiciones": "interactividad_exhibiciones",
    "23_limpieza_general": "limpieza_general",
    "24_sealizacin_y_orientacin": "senalizacion_orientacion",
    "25_limpieza_de_los_baos": "limpieza_banos",
    "42_valoracin_del_tiempo_empleado_en_la_visita": "valoracion_tiempo_visita",
    "43_cunto_tiempo_dedic_a_la_visita_en_minutos": "tiempo_visita_minutos"
}, inplace=True)

# Mostrar nombres corregidos
df_cleaned.head()



```

#### Manejar valores nulos solo en variables clave (rellenar con la media en numéricas,moda en categóricas)

```python
df_cleaned.isnull().sum()
```

```python
df_cleaned.info()
```

```python
# Las variables claves definidas no presentan valores nulos despues de filtrar status = Parcial
# Sin embargo, procedo a cumplir con las 3 columnas que presentan valores nulos.
```

```python
# facilidad_compra
df_cleaned['facilidad_compra'].fillna(df_cleaned['facilidad_compra'].mean(), inplace=True)

# interactividad_exhibiciones
df_cleaned['interactividad_exhibiciones'].fillna(df_cleaned['interactividad_exhibiciones'].mean(), inplace=True)

# limpieza_banos 
df_cleaned['limpieza_banos'].fillna(df_cleaned['limpieza_banos'].mean(), inplace=True)

df_cleaned.isnull().sum()
```

# 2. Exploración y cálculo de métricas clave 

### Obtener media y desviación estándar de satisfacción general

```python
satisfaccion_media = round(df_cleaned['satisfaccion_general'].mean(), 2)
satisfaccion_std = round(df_cleaned['satisfaccion_general'].std(), 2)

print(f"El promedio de satisfacción general es: {satisfaccion_media}")
print(f"La desviación estándar de satisfacción general es: {satisfaccion_std}")
```

### Contar el número de encuestas por tipo de visitante

```python
encuentas_tipo = df_cleaned['tipo_visitante'].value_counts()
encuentas_tipo
```

### Calcular correlación entre satisfacción general y facilidad de acceso

```python
df_numeric = df_cleaned.select_dtypes(include=['number'])

corr_matrix = df_numeric.corr()

correlacion = round(corr_matrix.loc['satisfaccion_general', 'facilidad_acceso'], 2)
print(f"La correlacion entre satisfacción general y facilidad de acceso es de : {correlacion}")
```

### Calcular el porcentaje de Promotores (9-10), Pasivos (7-8) y Detractores (0-6) a partir del Net Promoter Score (NPS)

```python
df_cleaned['nps_category'] = pd.cut(df_cleaned['nps'],
                                    bins=[0., 6, 8, 10],
                                    labels=['Detractores','Pasivos','Promotores'])

nps_counts = df_cleaned['nps_category'].value_counts()

total_respuestas = len(df_cleaned)
nps_percentages = (nps_counts / total_respuestas * 100).round(2)


print("Distribución del Net Promoter Score (NPS):")
print(nps_percentages)
```

# 3. Transformación y exportación de datos 

### Crear las siguientes tablas y exportarlas en CSV para Power BI: 
1. Resumen de Satisfacción: Promedio de satisfacción por tipo de visitante. 
2. Tabla de NPS: Distribución de respuestas segmentadas por tipo de visitante, con 
cálculo de % de Promotores, Pasivos y Detractores. 
3. Tiempo de visita vs. Satisfacción: Promedio del tiempo de visita en función del 
nivel de satisfacción. 
4. Frecuencia de respuestas: Número de encuestas completadas por tipo de 
visitante. 
5. Factores clave en la satisfacción: Media de cada criterio de evaluación. 
6. Impacto en la recomendación: Relación entre NPS y facilidad de compra, 
claridad del recorrido, etc. 

```python
resumen_satisfaccion = df_cleaned.groupby('tipo_visitante')['satisfaccion_general'].mean().reset_index()
resumen_satisfaccion
```

```python
distribucion_nps = df_cleaned.groupby(['tipo_visitante','nps_category'], observed=False).size().unstack(fill_value=0)

nps_percentage = distribucion_nps.div(distribucion_nps.sum(axis=1), axis=0) * 100
nps_percentage 
```

```python
tiempo_vs_satisfaccion = df_cleaned.groupby("satisfaccion_general")["tiempo_visita_minutos"].mean().reset_index()
tiempo_vs_satisfaccion
```

```python
frecuencia_respuestas = df_cleaned["tipo_visitante"].value_counts().reset_index()
frecuencia_respuestas.columns = ["Tipo de visitante", "Número de encuestas"]
frecuencia_respuestas
```

```python
factores_clave = df_cleaned[
    [
        "facilidad_compra",
        "facilidad_acceso",
        "claridad_recorrido",
        "fluidez_visita",
        "calidad_elementos",
        "variedad_elementos",
        "interactividad_exhibiciones",
        "limpieza_general",
        "senalizacion_orientacion",
        "limpieza_banos",
    ]
].mean().reset_index()
factores_clave.columns = ["Criterio", "Media de satisfacción"]
factores_clave
```

```python
impacto_recomendacion = df_cleaned.groupby("nps_category", observed=False)[["facilidad_compra", "claridad_recorrido", "limpieza_general"]].mean().reset_index()
impacto_recomendacion
```

```python
# Guardar los resultados en CSV
resumen_satisfaccion.to_csv("resumen_satisfaccion.csv", index=False, float_format="%.2f", decimal=",")
nps_percentage.to_csv("tabla_nps.csv",float_format="%.2f", decimal=",")
tiempo_vs_satisfaccion.to_csv("tiempo_vs_satisfaccion.csv", index=False, float_format="%.2f", decimal=",")
frecuencia_respuestas.to_csv("frecuencia_respuestas.csv", index=False, float_format="%.2f", decimal=",")
factores_clave.to_csv("factores_clave.csv", index=False, float_format="%.2f", decimal=",")
impacto_recomendacion.to_csv("impacto_recomendacion.csv", index=False, float_format="%.2f", decimal=",")
```
